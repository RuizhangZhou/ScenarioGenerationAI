```python
import torch
import pandas as pd
import os

# 输入图片所在文件夹
# 读取CSV文件
def read_csv_file(file_path):
    df = pd.read_csv(file_path)
    return df

def get_all_file_paths(folder_path):
    all_file_paths = []
    for root, dirs, files in os.walk(folder_path):
        for file in files:
            file_path = os.path.abspath(os.path.join(root, file))
            all_file_paths.append(file_path)
    return all_file_paths

folder_path = '/DATA1/rzhou/ika/testcases/inD/full_cases/18-29'  # 替换为你的文件夹路径
file_paths = get_all_file_paths(folder_path)
print(file_paths)


# 读取所有CSV文件
n = len(file_paths)  # 假设有n个文件
datasets = [read_csv_file(file_path) for file_path in file_paths]
```

```python
import pandas as pd

# 假设`datasets`是你的DataFrame
# 示例数据已经给出

# 检验x和y列是否包含非0元素
contains_non_zero_x = (datasets[0]['x'] != 0).any()
contains_non_zero_y = (datasets[0]['y'] != 0).any()

if contains_non_zero_x or contains_non_zero_y:
    print("存在非0元素")
else:
    print("x和y列全为0")
```

```python
dfs_final=[]
for dataset in datasets:
    df = pd.DataFrame(dataset)

    # 创建新的列名
    df['new_col'] = df['trackID'].astype(str)
    df['x_col'] = 'x' + df['new_col']
    df['y_col'] = 'y' + df['new_col']

    # 分别处理 x 和 y 坐标
    df_x = df.pivot(index='frame', columns='x_col', values='x').reset_index()
    df_y = df.pivot(index='frame', columns='y_col', values='y').reset_index()

    # 合并 x 和 y 数据
    df_merged = pd.merge(df_x, df_y, on='frame', how='outer')

    # 删除 frame 1500 的数据（如果存在）
    df_merged = df_merged[df_merged['frame'] != 1500]

    # 确保列顺序
    col_order = ['frame'] + [f'{coord}{i}' for i in range(1, 31) for coord in ('x', 'y')]
    df_final = df_merged.reindex(columns=col_order)
    dfs_final.append(df_final)
print(dfs_final)
```

```python
print(dfs_final[0])
```

```python
combined_df = pd.DataFrame()

# 遍历datasets中的每个DataFrame
for index, df in enumerate(dfs_final, start=1):
    # 给当前DataFrame添加caseID列，值为DataFrame的索引（即当前是第几个DataFrame）
    df['caseID'] = index
    # 将caseID列移动到DataFrame的第一列位置
    cols = df.columns.tolist()
    cols = cols[-1:] + cols[:-1]
    df = df[cols]
    # 将当前DataFrame添加到combined_df中
    combined_df = pd.concat([combined_df, df], ignore_index=True)

# 将合并后的DataFrame写入CSV文件
combined_df.drop('frame', axis=1, inplace=True)
combined_df.to_csv('/DATA1/rzhou/ika/multi_testcases/inD_multi_18-29_interval1500_v30_with0values.csv', index=False)

```
